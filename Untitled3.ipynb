```python

def get_words_from_lemmas(pos='n',lex='feeling'):
    noun_lemmas_in_wordnet = set(chain(*[ss.lemma_names() for ss in wn.all_synsets(pos=pos) if lex in ss.lemma_names()]))
    return noun_lemmas_in_wordnet
```
